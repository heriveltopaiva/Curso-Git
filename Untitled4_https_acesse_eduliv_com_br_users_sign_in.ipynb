Quando você atribui um valor ou um objeto a uma variável, o Python encontra um espaço na memória para armazenar esse objeto.
A variável que você nomeou aponta para aquele local na memória.
Você pode então "chamar" ou "reutilizar" esse objeto referenciando o nome da variável em qualquer lugar do seu código, subsequentemente, até que o programa termine de rodar ou você sobrescreva a variável com um novo valor.

Sim, na linguagem Python, **a atribuição de valor** a uma variável é **sempre** feita usando o sinal de igual (=).
Este operador é chamado de operador de atribuição.
O que o sinal de igual faz:
O sinal de = diz ao Python: **"Pegue o valor que está à minha direita e armazene-o na variável que está à minha esquerda".**

A variável torna o codigo mais eficiente a manutenção, por Ex:

**Código sem variáveis**

print(10 + 5)

print(10 * 2)

print("O valor é 10")

 Se o '10' precisar ser '15' depois, você tem que mudar 3 lugares.

 Com Variáveis (A Solução Eficiente)
Usando uma variável, você centraliza a informação. Você a define uma vez e a utiliza em múltiplas partes do código. Se o valor mudar, você altera apenas na linha onde a variável foi definida, e o restante do código se atualiza automaticamente:
python

** Código com variáveis**

valor_base = 10

print(valor_base + 5)

print(valor_base * 2)

print(f"O valor é {valor_base}")


 Se o valor precisar ser 15, mude SÓ AQUI:

valor_base = 15

 O restante do código abaixo usará automaticamente o novo valor 15.
Use o código com cuidado.

Portanto, seu raciocínio está correto. O uso de variáveis torna o código mais organizado, fácil de manter e flexível a mudanças.

Se for string estoiu juntando (resultado_soma = numero_1 + numero_2)

Se for numeros inteiros estou fazendo uma operação (resultado_soma = int(numero_1) + int(numero_2)?

Exatamente! Você entendeu perfeitamente a lógica do operador + em Python.
É um conceito-chave na linguagem, que podemos resumir da seguinte forma:
Se os dados são str (strings/textos), o + realiza uma concatenação (união, junção dos textos).
Se os dados são int (números inteiros) ou float (números decimais), o + realiza uma operação matemática de adição (soma).
